```python
# Function to find the index and column of a specific value in a data frame
def find_value(df, value):
    for col in df.columns:
        for idx in df.index:
            if df.at[idx, col] == value:
                return idx, col
    return None, None

# List of target values
target_values = [
    "Mobile Home or Trailer",
    "One-family house detached",
    "One-family house attached",
    "2 Apartments",
    "3-4 Apartments",
    "5-9 Apartments",
    "10-19 Apartments",
    "20-49 Apartments",
    "50 or More Apartments",
]

# Iterate over target_values
for target_value in target_values:
    index1, col1 = find_value(df1, target_value)

    # If the target value is found in df1, start merging with df2 and df3
    if index1 is not None and col1 is not None:
        # Iterate over the columns of df2
        for col2 in df2.columns:
            index2, _ = find_value(df2[col2].to_frame(), target_value)

            # If the target value is found in the current column of df2, insert it next to the matched column in df1
            if index2 is not None:
                new_col_name = f'OCRIP_{col2}'
                df1.insert(df1.columns.get_loc(col1) + 1, new_col_name, df2[col2])
                col1 = new_col_name  # Update col1 to insert the next match to the right

        # Iterate over the columns of df3
        for col3 in df3.columns:
            index3, _ = find_value(df3[col3].to_frame(), target_value)

            # If the target value is found in the current column of df3, insert it next to the matched column in df1
            if index3 is not None:
                new_col_name = f'GRPIP_{col3}'
                df1.insert(df1.columns.get_loc(col1) + 1, new_col_name, df3[col3])

df1.head()
```

```python
# set display option to show all columns
pd.set_option('display.max_columns', None)

# display the DataFrame with all columns
df1.head(8)
```

```python
# Remove rows 0 through 2
df1 = df1.drop([0, 1, 2], axis=0)

# Remove columns 1-4
df1 = df1.drop(df1.columns[1:5], axis=1)

df1.head(10)
```

```python
import numpy as np

# Group every 12 columns after the first one into separate DataFrames
categories = [
    "Mobile Home or Trailer",
    "One-family house detached",
    "One-family house attached",
    "2 Apartments",
    "3-4 Apartments",
    "5-9 Apartments",
    "10-19 Apartments",
    "20-49 Apartments",
    "50 or More Apartments",
]

dfs_by_category = {}
for i, category in enumerate(categories):
    start_col = 1 + i * 12
    end_col = start_col + 12
    dfs_by_category[category] = df1.iloc[:, start_col:end_col]

# Concatenate the processed DataFrames back into a single DataFrame with a MultiIndex
df1 = pd.concat([df1.iloc[:, :1]] + [dfs_by_category[category] for category in categories], axis=1)

# Create a MultiIndex with the category names as the top level
columns = pd.MultiIndex.from_product([categories, dfs_by_category[categories[0]].columns], names=['Category', 'Subcolumn'])

# Assign the new MultiIndex to df1
df1.columns = pd.MultiIndex.from_tuples([(np.nan, df1.columns[0])] + list(columns))

df1.head()

##
```

```python
# List of subcolumn names to be deleted for each category
subcolumns_to_delete = [
    "OCRIP_Unnamed: 21",
    "OCRIP_Unnamed: 31",
    "GRPIP_Unnamed: 31",
    "GRPIP_Unnamed: 11",
    "GRPIP_Unnamed: 1"
]

categories = [
    "Mobile Home or Trailer",
    "One-family house detached",
    "One-family house attached",
    "2 Apartments",
    "3-4 Apartments",
    "5-9 Apartments",
    "10-19 Apartments",
    "20-49 Apartments",
    "50 or More Apartments",
]

# Remove specified subcolumns from the multi-level DataFrame
df3 = df1.copy()
for category in categories:
    for subcolumn in subcolumns_to_delete:
        if (category, subcolumn) in df3.columns:
            df3 = df3.drop(columns=(category, subcolumn))

# Display the first 10 rows of the resulting DataFrame
df3.head(10)
```

```python

def is_number(value):
    try:
        float_value = float(value)
        return not pd.isna(float_value)
    except ValueError:
        return False

def safe_sum(a, b):
    a = int(float(a)) if is_number(a) else 0
    b = int(float(b)) if is_number(b) else 0
    return a + b

# Create a new DataFrame to store the processed data
df5 = df3.copy()

# Create the "Owned" and "Rented" subcolumns for each category by combining respective subcolumns
for category in categories:
    if (category, "Unnamed: 5") in df5.columns and (category, "Unnamed: 6") in df5.columns:
        df5[(category, "Owned")] = df5.apply(lambda row: safe_sum(row[(category, "Unnamed: 5")], row[(category, "Unnamed: 6")]), axis=1)
    if (category, "Unnamed: 7") in df5.columns and (category, "Unnamed: 8") in df5.columns:
        df5[(category, "Rented")] = df5.apply(lambda row: safe_sum(row[(category, "Unnamed: 7")], row[(category, "Unnamed: 8")]), axis=1)


# Reorder the columns based on the new order and sort by category
first_column_name = "Source: ACS 1-Year Estimates Public Use Microdata Sample 2021"
new_columns_order = [(np.nan, first_column_name)]
for category in categories:
    owned_col = (category, "Owned")
    rented_col = (category, "Rented")
    if owned_col in df5.columns:
        new_columns_order.append(owned_col)
    if rented_col in df5.columns:
        new_columns_order.append(rented_col)
    for subcol in df3[category].columns:
        if subcol not in ["Unnamed: 5", "Unnamed: 6", "Unnamed: 7", "Unnamed: 8"]:
            new_columns_order.append((category, subcol))

# Rename specific subcolumns
df5 = df5.rename(columns={
    "OCRIP_Unnamed: 1": "Monthly Owner Costs With Mortgage",
    "OCRIP_Unnamed: 11": "Monthly Owner Costs Without Mortgage",
    "GRPIP_Unnamed: 21": "Monthly Renter Costs"
}, level=1)

#Update new_columns_order with the new column names
new_columns_order = [('NaN', 'Source: ACS 1-Year Estimates Public Use Microdata Sample 2021')]  # Add the first column
for category in categories:
    new_columns_order.extend([
        (category, "Owned"),
        (category, "Rented"),
        (category, "Monthly Owner Costs With Mortgage"),
        (category, "Monthly Owner Costs Without Mortgage"),
        (category, "Monthly Renter Costs")
    ])

# Reorder the columns based on the new order and sort by category
df5 = df5.loc[:, new_columns_order]

# Display the first 10 rows of the resulting DataFrame
df5.head(10)

# Reorder the columns based on the new order and sort by category
df5 = df5.loc[:, new_columns_order]

# Display the first 10 rows of the resulting DataFrame
df5.head(10)

```
